```python
import cv2
import numpy as np
image= cv2.imread("lio2.jpeg")


model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces= model.detectMultiScale(image)
print(faces)

for face in faces:
  x,y,w,h = face
  cv2.rectangle(image, (x,y), (x+w,y+h), (255,0,0),3)
cv2.imwrite("lio2ans.jpg", image)#check why blue box not appearing
```
