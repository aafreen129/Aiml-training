```python
import numpy as np
import matplotlib.pyplot as plt
image=np.full((16,16,3),250,dtype=np.uint8)
image[4:10,4:5]=[255,0,0]
image[4:10,5:6]=[255,255,0]
plt.imshow(image,cmap='jet')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

image = np.full((18, 30, 3), 255, dtype=np.uint8)

image[0:6, :, :] = [255, 153, 51]

image[6:12, :, :] = [255, 255, 255]

image[12:18, :, :] = [19, 136, 8]

plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
image=np.full((160,160,3),250,dtype=np.uint8)
image[40:100,40:50]=[0,0,255]
image[40:100,40:50]=[255,0,0]
cv2.imshow('image',image)
cv2.imwrite('image.jpg',image)
cv2.imread('image.jpg')
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
image=cv2.imread('demo1.jpg')
print(type(image))

image=cv2.resize(image,(150,150))
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=cv2.flip(image,0)
r,b,g=cv2.split(image)
cv2.imshow('image',r)
print(image.shape)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
image=cv2.imread('demo1.jpg',0)

image=cv2.GaussianBlur(image, (5,5),0) 
threshold_image, th=cv2.threshold(image, 107,255, cv2.THRESH_TOZERO) 
cv2.imshow('image',th)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
